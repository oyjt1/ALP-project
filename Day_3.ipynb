```python
#save the file path to  a variable
import pandas as pd # import
melb_file_path = 'melb_data.csv'
#read the data and store data in dataframe title melb_data
melb_data = pd.read_csv(melb_file_path)
#print a  summary of data in melb_data
melb_data.head()
```

```python
melb_data = melb_data.dropna(axis=0)
```

selecting the Target prediction

```python
y = melb_data.Price
```

Choosing features

```python
melb_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'Lattitude', 'Longtitude']
```

```python
X = melb_data[melb_features]
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melb_model = DecisionTreeRegressor(random_state=1)

# Fit model
melb_model.fit(X, y)
```

MAE: Mean Absolute Error
ERRor =(actual value)-(predicted value)

```python
#Import the mean_aboslute_error function from sklearn.metrics
from sklearn.metrics import mean_absolute_error#

predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,val_x,train_y,val_y = train_test_split(X,y,random_state=0)
# we split the data into the traning part and the validation/evaluation part.
# we do this for both the target and the feature .
#since we use random_state= 0 this means that we don't jumble the data when we split it .

melb_model2= DecisionTreeRegressor()
melb_model2.fit(train_x,train_y)
val_predictions = melb_model2.predict(val_x)
print(mean_absolute_error(val_y,val_predictions))
```
